```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("compiled.csv", encoding='utf-8')
```

```python
data[['hour', 'hour_sin', 'hour_cos', 'month', 'month_sin', 'month_cos', 'date']].head(24)
```

```python
def encode(data, col, max_val):
    data[col + '_sin'] = np.sin(2 * np.pi * data[col]/max_val)
    data[col + '_cos'] = np.cos(2 * np.pi * data[col]/max_val)
    return data
```

```python
encode(data, 'hour', 24)
encode(data, 'month', 12)
encode(data, 'd ay', 31)
```

```python
filename = r"compiled2.csv"
data.to_csv(filename, encoding='utf-8', index=False, header=True)
```
